# Dimensional Transformation

### Reshape

```python
x=tf.range(96)
x=tf.reshape(x,[4,24])
print(x)
```

```python
x.ndim, x.shape
```

```python
tf.reshape(x,[2,48])
```

```python
tf.reshape(x,[8,12])
```

```python
tf.reshape(x,[2,4,12])
```

```python
tf.reshape(x,[2,-1,3])
```

### Add and delete dimensions

```python
tf.random.uniform([28,28],maxval=10, dtype=tf.int32)
```

```python
tf.expand_dims(x, axis=2)
```

```python
x=tf.expand_dims(x, axis=0)
x
```

```python
x=tf.squeeze(x, axis=0)
x
```

```python
x=tf.squeeze(x, axis=2)
x
```

```python
x=tf.random.uniform([1,28,28,1],maxval=10, dtype=tf.int32)
x
```

### Swap dimensions

```python
x=tf.random.normal([2,32,32,3])
tf.transpose(x, perm=[0,3,1,2])
```

```python
tf.transpose(x, perm=[0,2,1,3])
```

### Copy data

```python
b=tf.constant([1,2])
b=tf.expand_dims(b, axis=0)
b
```

```python
b=tf.tile(b, multiples=[2,1])
b
```

```python
x=tf.range(4)
x=tf.reshape(x, [2,2])
x
```

```python
tf.tile(x, multiples=[1,2])
```

```python
tf.tile(x, multiples=[2,1])
```
