# Indexing and Slicing

### Indexing

```python
x=tf.random.normal([4,32,32,3])
x[0]
```

```python
x[0][1]
```

```python
x[0][1][2]
```

```python
x[0][1][2][0]
```

### Slicing

```python
x[1:3]
```

```python
x[0,::] # read 1st picture
```

```python
x[:,0:28:2,0:28:2,:]
```

```python
x=tf.range(9)
x[8:0:-1]
```

```python
x[::-1]
```

```python
x[::-2]
```

```python
x=tf.random.normal([4,32,32,3])
x[0,::-2,::-2]
```

```python
x[:,:,:,1]
```

```python
x[0:2,...,1:]
```

```python
x[2:,...] # same as x[2:]
```

```python
x[...,:2]
```

# Mathematical Operations

```python
a=tf.range(5)
b=tf.constant(2)
```

### Addition

```python
a+b
```

### Subtraction

```python
a-b
```

### Multiplication

```python
a*b
```

### Division

```python
a/b
```

```python
a//b
```

```python
a%b
```

### Power

```python
x=tf.range(4)
tf.pow(x, 3)
```

```python
x**2
```

```python
x=tf.constant([1.,4.,9.])
x**(0.5)
```

```python
x=tf.range(5)
x=tf.cast(x, dtype=tf.float32)
x=tf.square(x)
x
```

```python
tf.sqrt(x)
```

### Exponential

```python
x=tf.constant([1.,2.,3.])
2**x
```

```python
tf.exp(1.)
```

### Logarithmic

```python
x=tf.exp(3.)
tf.math.log(x)
```

```python
x=tf.constant([1.,2.])
x=10**x
tf.math.log(x)/tf.math.log(10.)
```

### Matrix multiplication

```python
a=tf.random.normal([4,3,28,32])
b=tf.random.normal([4,3,32,3])
a@b # @ is matrix multiplication
```

```python
a=tf.random.normal([4,28,32])
b=tf.random.normal([32,16])
tf.matmul(a, b)
```
