# 9.7 Data Augmentation

```python
import tensorflow as tf
tf.__version__
```

```python
def preprocess(x,y):
    x = tf.io.read_file(x)
    x = tf.image.decode_jpeg(x, channels=3)
    x = tf.image.resize(x, [244, 244])
```

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
x="./flower.jpg"
y=0
preprocess(x,y)
image = mpimg.imread("./flower.jpg")
plt.imshow(image)
plt.show()
```

## 9.7.1 Rotation

```python
x=tf.image.rot90(image,2)
plt.imshow(x)
plt.show()
```

## 9.7.2 Flip

```python
x = tf.image.random_flip_left_right(image)
plt.imshow(x)
plt.show()
```

```python
x = tf.image.random_flip_up_down(image)
plt.imshow(x)
plt.show()
```

## 9.7.3 Cropping

```python
x = tf.image.random_crop(image, [600,600,3])
plt.imshow(x)
plt.show()
```
