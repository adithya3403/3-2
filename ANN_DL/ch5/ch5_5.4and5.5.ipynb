# 5.5 Data Limiting

```python
x = tf.range(9)
tf.maximum(x,2) # Set lower limit of x to 2
```

```python
tf.minimum(x,7)
```

```python
x = tf.range(9)
tf.minimum(tf.maximum(x,2),7)
```

```python
x = tf.range(9)
tf.clip_by_value(x,2,7)
```
