```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
tf.__version__
```

# 6.5 Design of Output Layer

## [0,1] Interval with sum 1

```python
z = tf.constant([2.,1.,0.1])
tf.nn.softmax(z)
```

```python
# draw softmax function
z = np.array([2,1,0.1])
y = tf.nn.softmax(z)
plt.figure(figsize=(4,4))
plt.plot(z,y)
plt.grid()
plt.show()
```

```python
z = tf.random.normal([2,10])
y_onehot = tf.constant([1,3])
y_onehot = tf.one_hot(y_onehot, depth=10)
loss = keras.losses.categorical_crossentropy(y_onehot,z,from_logits=True)
loss = tf.reduce_mean(loss)
loss
```

```python
criteon = keras.losses.CategoricalCrossentropy(from_logits=True)
loss = criteon(y_onehot,z)
loss
```

## (-1,1) interval

```python
x = tf.linspace(-6.,6.,10)
tf.tanh(x)
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,tf.tanh(x),label='tanh')
plt.plot(x,tf.sigmoid(x),label='sigmoid')
plt.plot(x,tf.nn.relu(x),label='relu')
plt.legend()
plt.grid()
plt.show()
```
